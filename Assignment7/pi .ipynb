```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt

# Implement a program to estimate π using the simulation method.
# (a) The program should generate n random points of (x,y) where 0 ≤ x < 1 and
#  0 ≤ y < 1 for n ∈ {10^3,10^4,10^5,10^6}.
# You can use math.pi to compute error rates.
# error(%) = | π | × 100
# For n = 104, draw a scatter plot as Figure 1, i.e., blue color for dots
# whose distances from the origin (0, 0) are less than 1, otherwise red color.
```

```python
def findCoordinates(num):
  inCircleXCoord = []
  inCircleYCoord = []
  outOfCircleXCoord = []
  outOfCircleYCoord = []
  
  for i in range(num):
    # randomPoints = np.random.rand(num, 2)
    x = np.random.random()
    y = np.random.random()

    # https://math.stackexchange.com/questions/198764/how-to-know-if-a-point-is-inside-a-circle
    distanceFormula = math.sqrt(x**2 + y**2)
    if distanceFormula < 1:
      inCircleXCoord.append(x)
      inCircleYCoord.append(y)
    else:  
      outOfCircleXCoord.append(x)
      outOfCircleYCoord.append(y)

  return inCircleXCoord, inCircleYCoord, outOfCircleXCoord, outOfCircleYCoord
    
```

```python
def calculateError(num, inCircleXCoord):
  # find pi formula : number of points inside the circle(x coords)/ # of points outside the circle * 4
  pi_estimation = (len(inCircleXCoord)/num) * 4

  # find error : |(math.pi - pi) * pi | * 100
  errorRate = abs((math.pi - pi_estimation) / math.pi)  * 100

  # print in format  --- n = 10000, pi = 3.159200, error = 0.5605 %
  print("n = {0:}, pi = {1:.6f}, error = {2:.5f}%".format(num, pi_estimation, errorRate))

  return pi_estimation, errorRate
```

```python
def plot(inCircleXCoord, inCircleYCoord, outOfCircleXCoord, outOfCircleYCoord, pi_estimation, errorRate):
  plt.scatter(inCircleXCoord, inCircleYCoord, 0.5, color = 'b', marker=".", label = 'Inside Circle')
  plt.scatter(outOfCircleXCoord, outOfCircleYCoord, 0.5, color='r',marker=".", label = 'Outside Circle')
  plt.title("Figure 1: Estimating π Using Simulation") 
  # plt.legend(loc = 'upper right')
  plt
```

```python
# not plotting 10^3, 10^5, 10^6
def showValues(size):
  # find the coordinates for the circle
  inCircleXCoord, inCircleYCoord, outOfCircleXCoord, outOfCircleYCoord = findCoordinates(size)

  # calculate the error rate
  pi_estimation, errorRate = calculateError(size, inCircleYCoord)
```

```python
# for 10^4 graph
def simulate(size):
  #find the coordinates for the circle
  inCircleXCoord, inCircleYCoord, outOfCircleXCoord, outOfCircleYCoord = findCoordinates(size)

  # calculate the error rate
  pi_estimation, errorRate = calculateError(size, inCircleYCoord)

    # plot
  plot(inCircleXCoord, inCircleYCoord, outOfCircleXCoord, outOfCircleYCoord, pi_estimation, errorRate)
```

```python
# driver
def main():
  showValues((10 ** 3))
  simulate((10 ** 4))
  showValues((10 ** 5))
  showValues((10 ** 6))
```
